```python
import pandas as pd
import plotly.express as px

# Load the data from the CSV file
data = pd.read_csv('P3HT_optical.csv')

# Ensure 'Absorbance' is numeric and clean the data
data['Absorbance'] = pd.to_numeric(data['Absorbance'], errors='coerce')
clean_data = data.dropna()
clean_data = clean_data[(clean_data['Absorbance'] > -0.01) & (clean_data['Absorbance'] < 0.1)]

# Create an interactive plot using Plotly Express
fig = px.line(clean_data, x='Wavelength', y='Absorbance', 
              title='Absorbance Spectrum of P3HT',
              labels={'Wavelength': 'Wavelength (nm)', 'Absorbance': 'Absorbance (a.u.)'})

# Show plot
fig.show()
```

```python
import pandas as pd
import plotly.graph_objs as go

# Load the data from the CSV file
data = pd.read_csv('P3HT_optical.csv')

# Ensure 'Absorbance' is numeric and clean the data
data['Absorbance'] = pd.to_numeric(data['Absorbance'], errors='coerce')
clean_data = data.dropna()
clean_data = clean_data[(clean_data['Absorbance'] > -0.01) & (clean_data['Absorbance'] < 0.1)]

# Sort the data by wavelength to ensure the plot is from low to high
clean_data = clean_data.sort_values(by='Wavelength', ascending=True)

# Create a plot using Plotly
fig = go.Figure()

# Add the absorbance data as a line
fig.add_trace(go.Scatter(x=clean_data['Wavelength'], y=clean_data['Absorbance'],
                         mode='lines',
                         name='Absorbance'))

# Update the layout for better visualization
fig.update_layout(
    title='Absorbance Spectrum of P3HT',
    xaxis_title='Wavelength (nm)',
    yaxis_title='Absorbance (a.u.)',
    xaxis=dict(range=[clean_data['Wavelength'].min(), clean_data['Wavelength'].max()]),
    template='plotly_white'
)

# Show plot
fig.show()
```

```python
import pandas as pd
import plotly.graph_objs as go

# Load the data from the CSV file
data = pd.read_csv('P3HT_optical.csv')

# Ensure 'Absorbance' is numeric and clean the data, keeping the appropriate wavelength range
data['Absorbance'] = pd.to_numeric(data['Absorbance'], errors='coerce')
clean_data = data.dropna()
clean_data = clean_data[(clean_data['Wavelength'] >= 350) & (clean_data['Wavelength'] <= 1200)]

# Sort the data by wavelength to ensure the plot is from low to high
clean_data = clean_data.sort_values(by='Wavelength', ascending=True)

# Create a plot using Plotly
fig = go.Figure()

# Add the absorbance data as a line
fig.add_trace(go.Scatter(x=clean_data['Wavelength'], y=clean_data['Absorbance'],
                         mode='lines',
                         name='Absorbance'))

# Update the layout for better visualization
fig.update_layout(
    title='Absorbance Spectrum of P3HT',
    xaxis_title='Wavelength (nm)',
    yaxis_title='Absorbance (a.u.)',
    xaxis=dict(
        tickmode='linear',
        tick0=350,
        dtick=50,
        range=[350, 1200]
    ),
    template='plotly_white'
)

# Adjust layout to avoid overlapping labels
fig.update_xaxes(tickangle=45)
fig.update_layout(margin=dict(l=40, r=40, t=40, b=40), title_x=0.5)

# Show plot
fig.show()
```
